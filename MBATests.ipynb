```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('MBA Tests.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(columns=['dob','gender','roll_no'],inplace=True)
```

```python
percent_cols = [col for col in df.columns if col.endswith('percentage')]

for col in percent_cols:
    df[col] = df[col].str.replace('%', '').astype(float)
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], errors='coerce',dayfirst=True)
df['submit_time'] = pd.to_datetime(df['submit_time'], errors='coerce',dayfirst=True)
```

```python
df['duration_minutes'] = (df['submit_time'] - df['start_time']).dt.total_seconds() / 60
```

```python
df['phone'] = df['phone'].fillna(0).astype(int).astype(str)
df.loc[df['phone'] == '0', 'phone'] = None
```

```python
df['Email'] = df['Email'].str.lower().str.strip()
```

```python
df.drop_duplicates(inplace=True)
```
