```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\LPU Assignment MBA (1)\MBA Tech\MBA Tokens.csv")
```

```python
df.head()
```

```python
print("Initial Shape:", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())
```

```python



original_columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Original Columns:\n")
for col in original_columns:
    print(col)

```

```python
df.drop_duplicates(inplace=True)
```

```python
threshold = 0.5 * len(df)
df = df.dropna(thresh=threshold, axis=1)
```

```python
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python

date_cols = [
    'token_date', 'lead_created_date', 'full_payment_date', 'offer_released_date',
    'start_of_the_week', 'test_submitted_date', 'first_call_date', 'application_date'
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)
```

```python
print("\nData Types:\n", df.dtypes)
```

```python
print("\nFinal Shape:", df.shape)
print(df.head())
```

```python

total_token_paid = df[df['token_paid'] == 1].shape[0]


enrolled_count = df[(df['token_paid'] == 1) & (df['Enrolled'] == 'Enroled')].shape[0]


conversion_rate = (enrolled_count / total_token_paid) * 100

print(f"🎯 Total Token Paid: {total_token_paid}")
print(f"✅ Enrolled After Token: {enrolled_count}")
print(f"📊 Conversion Rate: {conversion_rate:.2f}%")
```

```python

agent_enrollments = df[df['Enrolled'] == 'Enroled'].groupby('agent').size().sort_values(ascending=False)

agent_df = agent_enrollments.reset_index()
agent_df.columns = ['Agent Name', 'No. of Candidates Enrolled']

# Display
print(agent_df)
```

```python



city_enrollments = df[df['Enrolled'] == 'Enroled'].groupby('city').size().sort_values(ascending=False)

city_df = city_enrollments.reset_index()
city_df.columns = ['City', 'No. of Candidates Enrolled']


print(city_df)
```

```python

state_enrollments = df[df['Enrolled'] == 'Enroled'].groupby('state').size().sort_values(ascending=False)


state_df = state_enrollments.reset_index()
state_df.columns = ['State', 'No. of Candidates Enrolled']

print(state_df)
```

```python

total_leads = df.groupby('agent').size()
```

```python

enrolled_leads = df[df['Enrolled'] == 'Enroled'].groupby('agent').size()
```

```python

agent_conversion = pd.DataFrame({
    'Total Leads': total_leads,
    'Enrolled Leads': enrolled_leads
})

agent_conversion['Enrolled Leads'] = agent_conversion['Enrolled Leads'].fillna(0)

agent_conversion['Conversion Rate (%)'] = (agent_conversion['Enrolled Leads'] / agent_conversion['Total Leads']) * 100

agent_conversion = agent_conversion.sort_values(by='Conversion Rate (%)', ascending=False)

print(agent_conversion.reset_index())
```

```python

enrolled_df['Month'] = enrolled_df['full_payment_date'].dt.to_period('M')

monthly_trend = enrolled_df.groupby('Month').size().reset_index(name='Enrollments')

monthly_trend['Month'] = monthly_trend['Month'].astype(str)

print(monthly_trend)
```

```python

token_stage = df[df['token_paid'] == 1].shape[0]

test_stage = df[(df['token_paid'] == 1) & (df['test_submitted_date'].notna())].shape[0]

application_stage = df[(df['token_paid'] == 1) & (df['application_date'].notna())].shape[0]

enrolled_stage = df[(df['token_paid'] == 1) & (df['Enrolled'] == 'Enroled')].shape[0]

funnel_data = {
    "Token Paid": token_stage,
    "Test Submitted": test_stage,
    "Application Submitted": application_stage,
    "Enrolled": enrolled_stage
}


for stage, count in funnel_data.items():
    print(f"{stage}: {count}")
```

```python

token_stage = df[df['token_paid'] == 1].shape[0]
test_stage = df[(df['token_paid'] == 1) & (df['test_submitted_date'].notna())].shape[0]
application_stage = df[(df['token_paid'] == 1) & (df['application_date'].notna())].shape[0]
enrolled_stage = df[(df['token_paid'] == 1) & (df['Enrolled'] == 'Enroled')].shape[0]


conv_test = (test_stage / token_stage) * 100 if token_stage else 0
conv_app = (application_stage / test_stage) * 100 if test_stage else 0
conv_enroll = (enrolled_stage / application_stage) * 100 if application_stage else 0
conv_total = (enrolled_stage / token_stage) * 100 if token_stage else 0

print(f"🔁 Token → Test: {conv_test:.2f}%")
print(f"📝 Test → Application: {conv_app:.2f}%")
print(f"📥 Application → Enrolled: {conv_enroll:.2f}%")
print(f"🎓 Overall (Token → Enrolled): {conv_total:.2f}%")
```

```python

df['Enrolled'] = df['Enrolled'].str.strip().str.lower()

no_test = df[(df['token_paid'] == 1) & (df['test_submitted_date'].isna())]

no_application = df[
    (df['token_paid'] == 1) &
    (df['test_submitted_date'].notna()) &
    (df['application_date'].isna())
]

not_enrolled = df[
    (df['token_paid'] == 1) &
    (df['application_date'].notna()) &
    (df['Enrolled'] != 'enroled')
]
```
