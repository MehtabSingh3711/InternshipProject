```python
df.columns = df.columns.str.strip()  # cleaning up column names
df.dtypes
```

```python
df['submit_time'] = pd.to_datetime(df['submit_time'], errors='coerce')
df['start_time'] = pd.to_datetime(df['start_time'], errors='coerce')
```

```python
percent_cols = [col for col in df.columns if 'percentage' in col.lower()]
for col in percent_cols:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('%', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
missing = df.isnull().sum() / len(df)
high_missing_cols = missing[missing > 0.95].index.tolist()
high_missing_cols
```

```python
df_raw = df.copy()
```

```python
columns_to_drop = [
    'dob', 'gender', 'degree_name', 'roll_no',
    'department_name', 'batch',
    'total_max_score', 'math_foundation_max_score',
    'data_interpretation_max_score', 'tech_and_computing_fundamentals_max_score',
    'verbal_reasoning_max_score', 'logical_reasoning_max_score',
    'general_reasoning_max_score'
]
existing_cols_to_drop = [col for col in columns_to_drop if col in df.columns]
df.drop(columns=existing_cols_to_drop, inplace=True)
```

```python
print("New shape of dataset:",df.shape)
```

```python
null_percent = df.isnull().mean() * 100
```

```python
columns_to_drop = null_percent[(null_percent == 100) | ((null_percent >= 98) & (null_percent < 100))].index
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
null_percent_updated = df_cleaned.isnull().mean() * 100
cols_1356 = null_percent_updated.round(2)[null_percent_updated.round(2) == 13.56].index.tolist()
```

```python
import matplotlib.pyplot as plt

if cols_1356:
    aligned_missing_rows = df_cleaned[cols_1356].isnull().all(axis=1)

    if aligned_missing_rows.sum() / len(df_cleaned) >= 0.1356:
        # Drop rows where all of cols_1356 are missing
        df_cleaned = df_cleaned[~aligned_missing_rows]
    else:
        for col in cols_1356:
            if df_cleaned[col].dtype == 'object':
                # Fill categorical with mode
                mode_val = df_cleaned[col].mode().iloc[0]
                df_cleaned[col].fillna(mode_val, inplace=True)

            elif pd.api.types.is_numeric_dtype(df_cleaned[col]):
                # Plot histogram
                plt.figure()
                df_cleaned[col].dropna().hist(bins=30)
                plt.title(f'Distribution of {col}')
                plt.grid(False)
                plt.show()

                # Fill based on skewness
                if abs(df_cleaned[col].skew()) > 1:
                    df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
                else:
                    df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)

            elif pd.api.types.is_datetime64_any_dtype(df_cleaned[col]):
                # For datetime columns, fill with most frequent date (mode)
                mode_val = df_cleaned[col].mode().iloc[0]
                df_cleaned[col].fillna(mode_val, inplace=True)
```

```python
print("Final cleaned shape:", df_cleaned.shape)
print("\nRemaining missing values (%):")
print(df_cleaned.isnull().mean() * 100)
```
