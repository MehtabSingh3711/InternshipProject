```python
import pandas as pd

```

```python
# Load dataset
df = pd.read_csv("MBA Tech Campaign performance.csv")

```

```python
df.info
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, how='all')
```

```python
df['campaign_start_date'].fillna('Unknown')
```

```python
df['ad_set_name'] = df['ad_set_name'].fillna('N/A')
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates()
```

```python
cols_to_clean = ['total_spent', 'impressions', 'clicks', 'click_through_rate']
for col in cols_to_clean:
    df[col] = df[col].astype(str).str.replace(',', '').str.replace('%', '').astype(float)
```

```python
df.groupby('platform')['leads'].sum().sort_values(ascending=False)
```

```python
df['CTR_calc'] = (df['clicks'] / df['impressions']) * 100
```

```python
df.to_csv("cleaned_campaign_data.csv", index=False)
```
