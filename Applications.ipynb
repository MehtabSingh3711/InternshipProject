```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Applications.csv')
```

```python
df.drop(columns=['amount','notes','candidate_id','lead_number'], inplace=True)
```

```python
df['submission_date'] = pd.to_datetime(df['submission_date'], errors='coerce',dayfirst=True)
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], errors='coerce',dayfirst=True)
df['joining_date'] = pd.to_datetime(df['joining_date'], errors='coerce',dayfirst=True)
```

```python
def clean_percentage(val):
    if pd.isnull(val):
        return np.nan
    val = str(val).strip().lower()
    if '%' in val:
        try:
            return float(val.replace('%', '').strip())
        except:
            return np.nan
    if 'cg' in val:
        try:
            cgpa = float(val.replace('cg', '').replace('cgpa', '').strip())
            return round((cgpa / 10) * 100, 2)
        except:
            return np.nan
    if '*' in val or val in ['a', 'b', 'c', 'd', 'e', 'c*']:
        return np.nan
    try:
        return float(val)
    except:
        return np.nan
```

```python
for col in ['x_percentage', 'xii_percentage', 'bachelor_percentage']:
    df[col] = df[col].apply(clean_percentage)
```

```python
df[['x_percentage', 'xii_percentage', 'bachelor_percentage']].describe()
```

```python
df.loc[df['xii_percentage'] > 100, 'xii_percentage'] = df['xii_percentage'] / 10
```
