```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('MBA Tech Campaign performance.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(df.columns[14:24],axis = 1,inplace=True)
```

```python
print("DataFrame Information:")
display(df.info())
print("Shape of the DataFrame:")
display(df.shape)
```

```python
df.head()
```

```python
df['dates'] = pd.to_datetime(df['dates'], errors='coerce',dayfirst=True)
df['campaign_start_date'] = pd.to_datetime(df['campaign_start_date'], errors='coerce',dayfirst=True)
```

```python
df['total_spent'] = df['total_spent'].replace({',': ''}, regex=True).astype(float)
df['impressions'] = df['impressions'].replace({',': ''}, regex=True).astype('Int64')
df['clicks'] = df['clicks'].replace({',': ''}, regex=True).astype('Int64')
```

```python
df['click_through_rate'] = df['click_through_rate'].replace({'%': ''}, regex=True).astype(float)
```

```python
df['ad_set_id'] = df['ad_set_id'].astype('Int64')
```
